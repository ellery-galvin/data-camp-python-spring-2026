# Installing packages

## What is a package?
- Packages are bundles of code other people have written that you can use out the box.  
- For example, the `pandas` package provides capabilities for tabular data processing.
- Packages typically have many functions and objects you can refer to and use.
- The documentation for `pandas` is here: https://pandas.pydata.org/docs/


## What is a package manager?
- Anaconda, which we will install together tomorrow, is a distribution of software that includes python, some IDEs, and a package manager called `conda`.
- A package manager is a program that helps you install, update, and manage packages on your computer.
- Instead of going to various websites, downloading files, and following complicated installation instructions, you can use a package manager to do it all for you with a single command on the command line (see below)

## Why is a package manager helpful?  
1. Packages often have a complex network of dependencies.  (e.g. in order for pandas to work, you also need many other packages already installed, like `numpy`).  Package managers automatically make sure you have all the packages you need installed so that you can just ask for `pandas` and get everything else in one go.
2. Packages are updated frequently, and thus there are often many versions of the same package available.  Even worse than the nightmare of dependencies, sometimes different packages depend on different versions of the same package.  Package managers automatically figure out which versions of which packages you need to install so that everything works together.
3. The code you write for a project today, itself, depends on specific versions of the packages you used.  Sometimes, as packages as updated, they "deprecate" certain functionality, meaning they remove it or change how it works.  If in a later project of yours, you were to replace the old version of `pandas` with a new version, your code might break.  To handle this issue, package managers use *environments* to keep track of which versions of which packages you used for a specific project.


## What is an environment?
- Typically, whenever we start a new python project we create a conda environment for it.  The environment contains the version of python we'll use along with the specific versions of any packages needed.
- You should see (intro-python) shown in your terminal prompt.  This is an environment I created for this course and automatically installed on your behalf.
- `conda` helps us create these environments and remember them across all our projects.  As a result, your computer will typically have many different versions of python and packages installed, each collection for a different project.

## Creating an environment and installing packages
- Anaconda comes with a `base` environment that already contains most of the packages that scientific developers use, but it doesn't contain `pandas`.
- It may be tempting to "pollute" the base environment by just installing extra things whenever you need them.  This is not a good idea because it makes it hard to keep track of which versions of which packages you have installed.  Instead, we'll create a new environment for this project.
- Go to the command line and type `conda create --name first_env pandas` to create a new environment called `first_env` with the `pandas` package installed.  You can replace `first_env` with any name you like.
- In general, command line interfaces like `conda` use commands with a hierarchical structure.  The first word is the name of the program, the second word is the command, and any additional words are arguments to the command.  In this case, `conda` is the program, `create` is the command, `--name` is an argument called `name` (argument names are denoted with two dashes), `first_env` is the value for the `--name` argument, and `pandas` is an unnamed argument specifying which package to install.  You can add many at the same time separated by spaces.
- The documentation for `conda` is here: https://docs.conda.io/projects/conda/en/latest/commands.html.


## Activating and modifying an environment
- To use an environment, you have to *activate* it.  You can do this from the command line by typing `conda activate first_env` (or whatever you named your environment).  The command prompt will change to indicate which environment is active.
- To install additional packages in the active environment, you can use the command `conda install package_name`.
- When the environment is active, any python code you run from the command line will use the version of python and packages in that environment.
- To deactivate the environment, you can type `conda deactivate`.
- In some IDEs, there is a visual interface for selecting and managing environments.  In VS Code notebooks, you have to tell VS Code what environment to use for the notebook (it uses the word `kernel` instead of `environment`, but the difference isn't important for the moment). # For loops

Each column in the DataFrame object is itself a special object in Pandas called a Series.

Notice how the `pandas` `Series` object automatically did the concatenation operation elementwise, putting "Alfred" together with "Packer".  That's a feature of the `Series` object.  Regular old lists work differently (remember `all_names` extended the list to be 6 elements and put the three last names after the three first names), towards different purposes.   

If we want to apply the concatenation operator `+` to every element of these lists (broadly called iterables), then we can loop (also called iterate) through them one at a time with `for` loops

```
for element_name in iterable_name:
    do something with element_name
```

You can make `element_name` any name you like that's a valid variable name.  That variable is accessible inside the for loop, but not outside.

Every indented line thereafter is repeated on each element.
    
Let's try

````python
# Iterate over the first names
print(first)
for first_name in first:
    print(first_name)

# Sum the numbers from 0 to 9
total = 0

# The range function generates an iterable sequence of numbers
print(range(10))
# But it's not itself a list, so it doesn't display like one

# Since it's iterable, we can use it in a for loop
for x in range(10):
    total = total + x
print(total)

# Add elements to the end of a list
full = []
full.append("a")
full.append("b")
print(full)

# The zip function pairs up elements from two or more iterables
full_names  = zip(first, last)
print(full_names)
# The zip object is an "iterable", but it doesn't display as a list
# Unless we unfurl it with the list() function
print(list(full_names))



# Apply concatenation to pairs of elements from two lists
full = []
# Here, we use that the zip object is iterable in a for loop
for fname, lname in zip(first, last):
    full.append(fname + " " + lname)
print(full)


# List comprehensions allow you to write the code above more concisely
full = []
full = [ fname + " " + lname for fname, lname in zip(first, last)]
print(full)

# Note we have shown how to make the full list two different ways
````

That last example is called a list comprehension.  It's a form that makes it very easy to build up lists in creative ways.  Using them is the canonical example of what some have come to call *pythonic* programming because it looks so natural.

# Demonstration: Data pipelines
- Data pipelines are sequences of data processing steps that transform raw data into useful information.
- In pandas, we can create data pipelines by chaining together multiple methods on a DataFrame using the dot notation.
- Each DataFrame method returns a new DataFrame, which we can then call another method on, and so on.

````python
articles_df["Authors"]
````

````python
# This code cell is a demonstration.
# We will not discuss each function in detail, but you can
# learn about them in the pandas documentation.


# Goal: Find the author with the most publications


# Take our data
authors = ( 
    # The parentheses indicate this is a single expression continued over multiple lines
    # Take our data
    articles_df
    # Reassign the column called Authors
    # To the result of applying the split_authors function to each value in the Authors column
    .assign(Authors=articles_df["Authors"].apply(split_authors))
    # Create a new row for each author in the Authors lists
    .explode("Authors")
    # Group by author for the next operation
    .groupby("Authors")
    # Compute the number of rows in each group (i.e. num articles per author)
    .size()
    # Sort descending to get the most prolific authors at the top
    .sort_values(ascending=False)
)
print(authors.head(10))
````

# Demonstration: Automation
- Automation is the process of using code to perform repetitive tasks without human intervention.
- For example, maybe we are creating profile pages for all the authors in our dataset that show their publication history.
- There are over 2500 authors!  We would never want to write html code for each author by hand.
- But with Python and Pandas, we can write a script that generates the html code for each author automatically, saving us a ton of time and effort.

````python
# Demonstration: Automation

# Get our data into long format, with one author per row
articles_df_long = (
    # Take our dataset
    articles_df
    # Replace the "Authors" col with the result of split_authors
    .assign(Authors=articles_df["Authors"].apply(split_authors))
    # Make a new rew for each author in the list for each article
    .explode("Authors")
)

# Get the list of unique authors
authors = articles_df_long["Authors"].unique()

# For each author
for author in authors:
    # Get all their publications
    publications = articles_df_long[ articles_df_long["Authors"] == author ]
    # Export to html
    html = publications.to_html()
    # Create a file named after the author in the export directory
    with open(f"export/{author.replace(' ', '_')}.html", "w") as f:
        # Write the html to the file
        f.write(html)
````

- `toolz` a library for making python more like a funcitonal programming language.  It provides currying and advanced datastreaming / iterator tools.
- `sqlite3` a library for interfacing with SQLite databases in python.
